### In Memory Store

```python
from langgraph.store.memory import InMemoryStore
```

```python
store = InMemoryStore()
```

```python
user1_namespace = ("users", "Amandeep")
user2_namespace = ("users", "Rahul")
```

```python
# store data
store.put(
    user1_namespace,
    'profile',
    {
        'name' : 'Amandeep',
        'city' : 'Haridwar'
    }
)
store.put(
    user2_namespace,
    'profile',
    {
        'name' : 'Rahul',
        'city' : 'Roorkee'
    }
)
```

```python
# Get data using namespace and key
result = store.get(
    namespace = user1_namespace,
    key = 'profile'
)
result
```

```python
items = store.search(user1_namespace)
for item in items:
    print(item.key)
    print(item.value)
```

### Vector Store

```python
from langgraph.store.memory import InMemoryStore
from langchain_openai import OpenAIEmbeddings
store = InMemoryStore(
    index = {
        'embed' : OpenAIEmbeddings(model = 'text-embedding-3-small'),
        'dims' : 1536
    }
)
namespace = ('users', 'Amandeep')
```

```python
store.put(namespace = namespace, key = 'memory1', value = 'Rahul enjoys writing Python code')
store.put(namespace = namespace, key = 'memory2', value = 'Rahul goes to the gym everyday')
store.put(namespace = namespace, key = 'memory3', value = 'Rahul loves Italian food')
store.put(namespace = namespace, key = 'memory4', value = 'Rahul is planning a vacation to Switzerland')
```
