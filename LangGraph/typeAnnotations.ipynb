```python
from typing import TypedDict, List # it will not enforce the rules
```

```python
class Student(TypedDict):
    name : str
    year : int
    marks : List[int]

stu = Student(
    name = "Amandeep",
    year = 2004,
    marks = [10, 20, 30]
)
print(stu)
print(type(stu))
```
