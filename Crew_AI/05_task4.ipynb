```python
from crewai import Agent, Task, Crew, LLM
from crewai.tools import BaseTool
from crewai_tools import TavilySearchTool
from dotenv import load_dotenv
import os
load_dotenv()
```

```python
llm = LLM(
    model="llama-3.3-70b-versatile",
    api_key=os.getenv("GROK_API_KEY"),
    base_url="https://api.groq.com/openai/v1"
)
```

```python
search_tool = TavilySearchTool()
```

```python
agent = Agent(
    role="Professional Blog Writer",

    goal="Research the given topic using Tavily Search and write an informative blog.",

    backstory="""
    You are an expert technical blog writer.
    Before writing any blog, always search the internet using the Tavily Search Tool
    to collect the latest and most accurate information.
    """,

    llm=llm,

    tools=[search_tool],

    verbose=True
)
```

```python
topic = """
Virat Kohli
"""
```

```python
task = Task(
    description="""
    Write a detailed blog on the topic:
    {topic}

    First use the Tavily Search Tool to gather relevant information.
    Then write a well-structured blog using the retrieved information.
    """,

    expected_output="""
    A professional blog with:
    - Title
    - Introduction
    - Main Content
    - Conclusion
    """,

    agent=agent
)
```

```python
crew = Crew(
    agents=[agent],
    tasks=[task],
    verbose=True
)
```

```python
result = await crew.kickoff_async(
    inputs={
        "topic": topic
    }
)
```

```python
from IPython.display import Markdown, display

display(Markdown(result.raw))
```

### Temperature

```python
from crewai.tools import BaseTool
import requests
import os

class WeatherTool(BaseTool):
    name: str = "Weather Tool"
    description: str = "Returns the current temperature of a given city."

    def _run(self, city: str) -> str:
        api_key = os.getenv("OPENWEATHER_API_KEY")

        url = "https://api.openweathermap.org/data/2.5/weather"

        params = {
            "q": city,
            "appid": api_key,
            "units": "metric"
        }

        response = requests.get(url, params=params)

        if response.status_code == 200:
            data = response.json()

            temp = data["main"]["temp"]
            weather = data["weather"][0]["description"]

            return f"The current temperature in {city} is {temp}°C with {weather}."

        return "Unable to fetch weather information."
```

```python
weather_tool = WeatherTool()
```

```python
agent = Agent(
    role="Weather Assistant",
    goal="Answer weather-related questions.",
    backstory="You can fetch live weather information using the Weather Tool.",
    llm=llm,
    tools=[weather_tool],
    verbose=True
)
```

```python
task = Task(
    description="Tell me the current temperature of Delhi.",
    expected_output="Current temperature of Delhi.",
    agent=agent
)
```

```python
crew = Crew(
    agents=[agent],
    tasks=[task],
    verbose=True
)
```
