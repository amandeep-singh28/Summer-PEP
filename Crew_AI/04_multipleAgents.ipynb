```python
from crewai import Agent, Task, Crew, LLM
from crewai.tools import BaseTool
from dotenv import load_dotenv
import os
load_dotenv()
```

```python
llm = LLM(
    model="llama-3.3-70b-versatile",
    api_key=os.getenv("GROK_API_KEY"),
    base_url="https://api.groq.com/openai/v1"
)
```

```python
agent1 = Agent (
    role = "You are an interesting fact finder",
    goal = "Give 5 different interesting facts",
    backstory = (
        "You are an expert interesting fact finder who finds interesting facts related to the topic given"
    ),
    llm = llm,
    verbose = True
)
```

```python
topic = """
Indian Cricket Team
"""
```

```python
task1 = Task (
    description = """
    Give 5 different interesting facts related to this {topic}
    """,
    expected_output = "Interesting Facts",
    agent = agent1
)
```

```python
agent2 = Agent (
    role = "You are a great blog writer",
    goal = "Create blog on the topic",
    backstory = (
        "You are an expert blog writer who writes blog to that particular topic"
    ),
    llm = llm,
    verbose = True
)
```

```python
task2 = Task (
    description = """
    Write a blog on this {topic}
    """,
    expected_output = "Blog",
    agent = agent2
)
```

```python
crew = Crew(
    agents = [agent1, agent2],
    tasks = [task1, task2],
    verbose = True
)
```

```python
result = await crew.kickoff_async(
    inputs = {
        "topic" : topic
    }
)
```
