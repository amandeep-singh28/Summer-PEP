```python
from crewai import Agent, Task, Crew, LLM
from crewai.tools import BaseTool
from dotenv import load_dotenv
import os
load_dotenv()
```

```python
llm = LLM(
    model="llama-3.3-70b-versatile",
    api_key=os.getenv("GROK_API_KEY"),
    base_url="https://api.groq.com/openai/v1"
)
```

```python
email = """
hey team,

just wanted to give a quick update.

the TAS stuff is mostly done now. i also fixed the issues we found in the PRX meeting.

i'll send the final docs after the DAC approves everything tomorrow.

let me know if i missed anything.

thanks
"""
```

```python
class JargonTool(BaseTool):
    name : str = "Company Jargon Tool"
    description : str = "Expands company abbreviations"

    def _run(self, text : str) -> str:
        mapping = {
            "TAS" : "Technical Architecture Stack",
            "PRX" : "Project Phoenix",
            "DAC" : "Design Approval Committee"
        }
        for k, v in mapping.items():
            text = text.replace(k, v)
        return text
tool = JargonTool()
```

```python
class AdditionTool(BaseTool):
    name : str = "Addition Tool"
    description : str = "This tool will add two numbers and returns the result"
    def _run(self, a : int, b : int):
        return a + b
tool1 = AdditionTool()
print(tool1.run(a=10, b=20))
```

```python
email_agent = Agent(
    role = "Professional Email Writer",
    goal = "Rewrite emails in a professional tone",
    backstory = (
        "You are an expert business communication specialist who writes clear, polite, and professional emails"
    ),
    llm = llm,
    tool = [tool],
    verbose = True
)
```

```python
task = Task(
    description = """
    Rewrite the following email professionally.
    Expand abbreviations if possible

    Email:
    {email}
    """,
    expected_output = "A polished professional email",
    agent = email_agent
)
```

```python
crew = Crew(
    agents = [email_agent],
    tasks = [task],
    verbose = True
)
```

```python
result = await crew.kickoff_async(
    inputs = {
        "email" : email
    }
)
```
