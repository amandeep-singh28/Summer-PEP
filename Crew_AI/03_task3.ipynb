```python
from crewai import Agent, Task, Crew, LLM
from crewai.tools import BaseTool
from dotenv import load_dotenv
import os
load_dotenv()
```

```python
llm = LLM(
    model="llama-3.3-70b-versatile",
    api_key=os.getenv("GROK_API_KEY"),
    base_url="https://api.groq.com/openai/v1"
)
```

```python
agent = Agent(
    role = "Your work is to only greet the user no matter what the prompt is",
    goal = "Greet the user",
    backstory = (
        "You are a assistant whose work is to greet the user"
    ),
    llm = llm,
    verbose = True
)
```

```python
text = input("Enter prompt?")
```

```python
task = Task(
    description = """
    Greet the user no matter the question is

    Message:
    {text}
    """,
    expected_output = "Greet the user",
    agent = agent
)
```

```python
crew = Crew(
    agents = [agent],
    tasks = [task],
    verbose = True
)
```

```python
result = await crew.kickoff_async(
    inputs = {
        "text" : text
    }
)
```

```python
print(result)
```
