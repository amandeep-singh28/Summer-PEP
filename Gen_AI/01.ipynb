```python
from dotenv import load_dotenv
import os

load_dotenv()
key = os.getenv("OPENAI_API_KEY")
gemini_key = os.getenv("GEMINI_API_KEY")
```

```python
# with openai
# if we have to switch the provider, than we need to use the other code
from openai import OpenAI
client = OpenAI(api_key = key)
response = client.responses.create(
    model = "gpt-5.5",
    input = "Write a one-sentence bedtime story about a unicorn."
)

print(response.output_text)
```

```python
# with langchain
# but in langchain we can use model parameter, we can directly make changes in it to use another model
from langchain_openai import ChatOpenAI
model = ChatOpenAI(
    model = "gpt-5.5",
    api_key = key,
    # max_completion_tokens = 100,
    temperature = 0.2
)
# prompt = "Write a one-sentence bedtime story about a unicorn."
prompt = "Write a poem about LPU."
result = model.invoke(prompt)
print(result.content)
```

```python
# gemini model
from langchain_google_genai import ChatGoogleGenerativeAI
model = ChatGoogleGenerativeAI(
    model = "gemini-3.5-flash",
    api_key = gemini_key
)
prompt = "Write a one-sentence bedtime story about a unicorn."
result = model.invoke(prompt)
print(result.content[0]["text"])
```

```python
from langchain_openai import OpenAIEmbeddings
embeddings = OpenAIEmbeddings(
    model = "text-embedding-3-small",
    api_key = key,
    dimensions = 500
)
cities_info = [
    "New Delhi is the capital city of India.",
    "Mumbai is known as the financial capital of India.",
    "Bengaluru is famous for its IT industry.",
    "Chennai is renowned for its rich cultural heritage.",
    "Kolkata is known as the City of Joy."
]
result = embeddings.embed_documents(cities_info)
print(result)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

query = "Amandeep"

query_embed = embeddings.embed_query(query)

similarities = cosine_similarity(
    [query_embed],
    result
)
print(similarities)
best_index = similarities.argmax()

print("Most Similar Document:")
print(cities_info[best_index])

print("\nSimilarity Score:")
print(similarities[0][best_index])
```
