```python
from langchain_openai import ChatOpenAI
from langchain_tavily import TavilySearch
```

```python
from dotenv import load_dotenv
import os

load_dotenv()

print(os.getenv("TAVILY_API_KEY"))
```

```python
llm = ChatOpenAI(
    model="gpt-4.1",
    temperature=0
)   
```

```python
# Create Tool
search_tool = TavilySearch(max_results=2)
```

```python
# Bind Tool
llm_with_tools = llm.bind_tools([search_tool]) # we pass an array of tools which we want to use
```

```python
# Tool Calling
response = llm_with_tools.invoke(
    "Tell me the todays latest news"
    # "What is the capital of India?"
)
```

```python
# earlier we were getting content part but now it is missing
# why? Because LLM can't produce the output right now, because the core task was to create the output using the tool not with the LLM.
print(response)
```

```python
print(response.tool_calls)
```

```python
# Tool Execution
tool_call = response.tool_calls[0]
```

```python
tool_call
```

```python
result = search_tool.invoke(tool_call["args"])

print(result)
```
